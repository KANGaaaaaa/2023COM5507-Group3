```python
import pandas as pd
import regex
import os

# Load the data
data = pd.read_csv('/Users/zwl77/5507/raw data/CNN comments.csv', encoding='utf-8')

# Delete irrelevant or duplicate columns
unnecessary_columns = [col for col in data.columns if 'Unnamed' in col]
data.drop(columns=unnecessary_columns, inplace=True)


# Convert date format
data['Publish Date'] = pd.to_datetime(data['Publish Date'], errors='coerce')

# Text cleaning
# Remove the corrupted symbol (e.g., �)
data['Comment'] = data['Comment'].astype(str).str.replace(r'\uFFFD', '', regex=True)

# Remove specific patterns: time stamps, dates, "TRUE", "FALSE"
data['Comment'] = data['Comment'].apply(lambda x: '' if isinstance(x, str) and (
    regex.fullmatch(r'\d{1,2}:\d{2}', x.strip()) or  
    regex.fullmatch(r'\d{1,2}月\d{1,2}日', x.strip()) or 
    regex.fullmatch(r'\d{4}/\d{1,2}/\d{1,2}', x.strip()) or  
    x in ["TRUE", "FALSE"]  
) else x)

import regex

data['Comment'] = data['Comment'].apply(lambda x: '' if isinstance(x, str) and (
    regex.fullmatch(r'^[\p{P}\s\d]*$', x.strip())  
) else x)


# Remove white space or comments containing only spaces
def clean_comment(comment):
    if isinstance(comment, str):
        comment = comment.strip() 
        if regex.fullmatch(r'^[\p{P}\s\d]*$', comment): 
            return ''  
    return comment  


data['Comment'] = data['Comment'].apply(clean_comment)

# Delete blank comments
data = data[data['Comment'] != '']

# Save the cleaned data to a new CSV file
cleaned_file_path = '/Users/zwl77/5507/raw data/CNN clean.csv'



data.to_csv(cleaned_file_path, index=False, encoding='utf-8')

# Provide the path to the cleaned file
cleaned_file_path
```
